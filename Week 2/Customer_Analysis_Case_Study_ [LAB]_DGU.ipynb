```python
def load_original_data():
    file1upt = pd.read_csv('Data/file1upt.csv')
    file2upt = pd.read_csv('Data/file2upt.csv')
    file3 = pd.read_csv('Data/file3.csv')

    return pd.concat([file1upt,file2upt,file3], axis=0)
```

```python
ca_df = load_original_data()
```

```python
ca_df
```

```python
ca_df = ca_df.drop_duplicates()
```

```python
ca_df
```

```python
#We lower all the letter in the header column with the function str.lower()
ca_df.columns= ca_df.columns.str.lower()
```

```python
ca_df.drop('customer', inplace=True, axis=1)
```

```python
#we will operate the lambda into isinstance() function to remove the "%" by a "" if it´s else than float or int
ca_df['customer lifetime value'] = ca_df['customer lifetime value'].apply(lambda x: x if isinstance(x, (float, int)) else x.replace("%", ""))
```

```python
#We will now use the same operation to remove the "/" before and after the 0.
ca_df["number of open complaints"] = ca_df["number of open complaints"].apply(lambda x: x[2] if isinstance(x, str) else x)
```

```python
#Now we will transform the types into float to be able to mean them
ca_df['customer lifetime value'] = ca_df['customer lifetime value'].astype(float)
#We change them finally in into integers
ca_df['customer lifetime value'] = ca_df['customer lifetime value'].astype(int)
```

```python
#creating a function to replace the different spelling of the genders to one unique per gender.
def clean_gender(x):
    if x in ['M', 'MALE',"male"]:
        return 'Male'
    elif x in ['F', 'female',"male"]:
        return 'Female'
    elif np.nan:  pass
    else:
        return 'U'
```

```python
#applying it
ca_df['gender'] = list(map(clean_gender, ca_df['gender']))
```

```python
#For the state we first want to know how much unique spelling they are 
print(ca_df['state'].unique())
```

```python
#creating a function to replace the different spelling of the States to a unique one.
def clean_state(x):
    if x in ['Cali']:
        return 'California'
    elif x in ['AZ']:
        return 'Arizona'
    elif x in ['WA']:
        return 'Washington'
    else:
        return x
```

```python
# We apply the operation
ca_df['state'] = list(map(clean_state, ca_df['state']))
```
