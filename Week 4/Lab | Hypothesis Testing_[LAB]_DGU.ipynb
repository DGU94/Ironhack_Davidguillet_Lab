Let's fix our confidence level to 95% which is the same as saying $\alpha = 1 - 0.95 = 0.05$

Hypothesis :
We are trying to refute the hypothesis that the population mean systolic blood pressure is μ = 120 mm Hg.

H0 = 120

H1 = 130.1

In order to determine the p-value, you **first need to determine the z-score**. To get the z-score, you can either reference the Z table or use scipy.stats in python to calculate it directly.

In SciPy, the z-value can be calculated with: stats.norm.cdf(z)

We use normal distribution! as the sample is > than 30.


Process :
Fix the confidence level. = 95 %
Compute the test statistic. My sample parameter  z 
Compute the critical value of the corresponding distribution.
Compare the test statistic with the critical value.
If the test statistic is in the allowed region, (the allowed region is determined by the confidence level) fail to reject  the null hypothesis, otherwise reject the null hypothesis.
The allowed region depends on the type of test(one tail-test or two-tail test).
